```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
file = pd.read_excel('bikedata.xlsx')
```

```python
file
```

### DATA CLEANING

Removing Duplicate

```python
file.drop_duplicates(inplace = True)
```

```python
file
```

```python
#Replacing 'M' & 'S' with 'Married' & 'Single' respectively

file['Marital Status']=file['Marital Status'].replace(['M','S'],['Married','Single'])
```

```python
file
```

```python
#Replacing 'M' & 'F' with 'Male' & 'Female' respectively

file['Gender']=file['Gender'].replace(['M','F'],['Male','Female'])
file
```

```python
#Grouping the ages in different age brackets

def age_grouping(age):
    if age<=30:
        return 'Adolescent'
    elif age<=54:
        return 'Middle Age'
    else:
        return 'old'
    
file['Age Bracket']=file['Age'].apply(age_grouping)
```

```python
file
```

### Showing relationships using Pivot table


```python
#Relationship between average income of each gender and their bike purchase history

piv1 = np.round(file.pivot_table (index= 'Gender',
                       columns= 'Purchased Bike',
                       values= 'Income',
                       aggfunc= 'mean'),0)
piv1
```

```python
#Relationship between distance travelled by cummuters and bike purchase

piv2= np.round(file.pivot_table (index= 'Commute Distance',
                                 values= 'Purchased Bike',
                                 aggfunc= 'count'),0)
piv2
```

```python
#Relationship between age bracket and bike purchase

piv3= np.round(file.pivot_table(index= 'Age Bracket',
                      values= 'Purchased Bike',
                      aggfunc= 'count'),0)
piv3
```

```python
#For Pivot table 1

piv1.plot(kind='bar', figsize=(8,5))
plt.title('Average Income by Gender and Bike Purchase')
plt.xlabel('Purchased Bike')
plt.ylabel('Average Income')
```

```python
#For Pivot table 2

piv2.plot(kind='line', figsize=(8,5))
plt.title('Bike Purchase by Commute Distance')
plt.xlabel('Commute Distance')
plt.ylabel('Bike Purchase')
```

```python
#For Pivot table 3

piv3.plot(kind='bar', figsize=(8,5))
plt.title('Bike Purchases by Age Bracket')
plt.xlabel('Age Bracket')
plt.ylabel('Bike Purchases')
```
